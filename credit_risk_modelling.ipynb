CREATE A FUNCTION FOR ALL OF THESE CALCULATIONS

```python
def woe_discrete(input_df, variable_name, good_bad_target_df):
    # JOIN INPUT DF OF VARIABLE AND TARGET DF
    df = pd.concat([input_df[variable_name], good_bad_target_df], axis=1)
    # GET THE COUNT AND MEAN OF VARIABLE WITH RESPECT TO TARGET
    df = pd.concat([
    df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].count(),
    df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].mean()
], axis=1)
    # REMOVE REPEATED COLUMNS
    df = df.iloc[:, [0, 1, 3]]
    # RENAME COLS
    df.columns = [df.columns.values[0], 'n_obs', '%_good']
    # GET % OF OBSERVATION RESPECTIVE TO EACH CATEGORY
    df['%_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    # GET THE COUNT OF GOOD AND BAD FOR A PARTICULAR CATEGORY
    df['n_good'] = df['%_good'] * df['n_obs']
    df['n_bad'] = (1 - df['%_good']) * df['n_obs']
    # OUT OF ALL GOOD/BAD, GET THE % FOR THE PARTICULAR CATEGORY OF GOOD/BAD
    df['%_n_good'] = df['n_good'] / df['n_good'].sum()
    df['%_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    # GET THE WoE
    df['WoE'] = np.log(df['%_n_good']/df['%_n_bad'])
    # SORT BY THE WoE THEN RESET THE INDEX, USE DROP = TRUE SINCE IF DONT SET IT, THE OLD INDEX WILL BE ADDED AS A NEW COLUMN
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop=True)
    # GET INFORMATION VALUE, SUM OF (%_n_good - $_n_bad) * WoE
    # IV IS THE SAME FOR ALL OF THE ROWS SINCE IT IS THE INFORMATION VALUE OF THE CATEGORY ITSELF
    df['IV'] = ((df['%_n_good'] - df['%_n_bad']) * df_1['WoE']).sum()
    return df
```
